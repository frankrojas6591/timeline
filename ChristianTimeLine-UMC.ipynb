```python
import pandas as pd
import numpy as np
from pathlib import Path
url_umcHistory = 'http://www.gcah.org/history/united-methodist-church-timeline'
df = pd.read_html(url_umcHistory, header=1)[0]
df.columns = ['UMC', 'Year', 'US']
df.Year = df.Year.fillna(0).astype(int)
df['cls'] = 'ev'
eList = []
for r in range(len(df)):
    rec = df.loc[r]
    print("======\n",rec)
    eIn = input("Enter class")
    if eIn == 'q' : break
    eList.append( rec.cls if eIn == '' else eIn)
df.ev = eList
df
    
    
```

```python
df.cls = eList
```

```python
umcDF = df[df.cls != 'ev'][['Year','cls','UMC']]
umcDF.to_csv('umcTimeline_gcah.csv')
print(len(umcDF))
umcDF
```
